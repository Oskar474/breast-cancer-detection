```python
import pandas as pd
df = pd.read_csv('data/breast-cancer.csv')
```

Wszystkie kolumny oprócz id oraz diagnosis są ciągłe

```python
import seaborn as sns
from matplotlib import pyplot as plt

df.drop('id', axis=1, inplace=True) 
df['diagnosis'] = (df['diagnosis'] == 'M').astype(int)

corr = df.corr()

plt.figure(figsize=(20,20))
sns.heatmap(corr, cmap=sns.color_palette("ch:s=-.2,r=.6", as_cmap=True),annot=True)
plt.show()
```

cechy która na macierzy posiadają wartość zbliżoną do 1 mają największą korelacje z diagnozą, zdecydowałem odrzucić cechy poniżej 0.2
